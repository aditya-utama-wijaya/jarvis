```python
from warnings import filterwarnings
filterwarnings('ignore')

import pandas as pd
pd.set_option('display.max_columns', None)

import numpy as np
from math import sqrt
from datetime import datetime
from keras.models import Sequential
from keras.layers import Dense, LSTM
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error as mse
import matplotlib.pyplot as plt
%matplotlib inline

path = 'data/'
```

```python
df = pd.read_csv(path + 'bitmex_orderbooks.csv', header = None)
print(df.shape)
pd.concat([df.head(), df.tail()])
```

```python
cols = [3, 23]
for c in cols:
    df[c] = [float(x[7:]) for x in df[c]]
df.tail()
```

```python
cols = list(range(4, 21, 2)) + list(range(24, 41, 2))
for c in cols:
    df[c] = [int(x[:-1]) for x in df[c]]
df.tail()
```

```python
cols = list(range(5, 22, 2)) + list(range(25, 42, 2))
for c in cols:
    df[c] = [float(x[1:]) for x in df[c]]
df.tail()
```

```python
cols = [22, 42]
for c in cols:
    df[c] = [int(x[:-2]) for x in df[c]]
df.tail()
```

```python
tmp = df.loc[:, 3:22]
tmp2 = tmp.stack().reset_index()[[0]]
tmp3 = pd.DataFrame({'bid_price': np.squeeze(tmp2[0::2].values), 'bid_volume': np.squeeze(tmp2[1::2].values)})
print(tmp3.shape)
pd.concat([tmp3.head(), tmp3.tail()])
```

```python
tmp = df.loc[:, 23:42]
tmp2 = tmp.stack().reset_index()[[0]]
tmp3['ask_price'] = np.squeeze(tmp2[0::2].values)
tmp3['ask_volume'] = np.squeeze(tmp2[1::2].values)
print(tmp3.shape)
pd.concat([tmp3.head(), tmp3.tail()])
```

```python
tmp = df[[43]]
tmp2 = tmp.loc[tmp.index.repeat(tmp3.shape[0] // df.shape[0])].reset_index(drop = True)
pd.concat([tmp2.head(), tmp2.tail()])
```

```python
tmp3['timestamp'] = tmp2[[43]]
tmp3['timestamp'] = [x[20:43] for x in tmp3['timestamp']]
tmp3['timestamp'] = [x.replace('T', ' ') for x in tmp3['timestamp']]
tmp3['timestamp'] = [x[:-4] for x in tmp3['timestamp']]
print(tmp3.shape)
tmp3[:30]
```

```python
tmp4 = tmp3.iloc[::10, :].reset_index(drop = True)
pd.concat([tmp4.head(), tmp4.tail()])
```

```python
tmp4['timetmp'] = [(x[8:10] + x[11:13] + x[14:16] + x[17:19]) for x in tmp4['timestamp']]
tmp5 = tmp4.groupby('timetmp').agg({'ask_price': np.mean}).reset_index()
print(tmp5.shape)
tmp5.tail()
```

```python
tmp5['timetmp'] = tmp5['timetmp'].astype('int')
print(min(tmp5['timetmp']), max(tmp5['timetmp']))
```

```python
timestamp = pd.to_datetime(tmp3['timestamp'], format = '%Y-%m-%d %H:%M:%S')
timestamp
```

```python
timestamp = pd.DataFrame({'timetmp': pd.date_range(timestamp[0], timestamp[len(timestamp) - 1], freq = '1S')})
timestamp['timetmp'] = [datetime.strftime(x, '%Y-%m-%d %H:%M:%S') for x in timestamp['timetmp']]
timestamp['timetmp'] = [int(x[8:10] + x[11:13] + x[14:16] + x[17:19]) for x in timestamp['timetmp']]
print(timestamp.shape)
pd.concat([timestamp.head(), timestamp.tail()])
```

```python
tmp6 = pd.merge(timestamp, tmp5, how = 'left', on = 'timetmp').fillna(method = 'ffill')
print(tmp6.shape)
pd.concat([tmp6.head(), tmp6.tail()])
```

```python
tmp6.drop('timetmp', axis = 1, inplace = True)
tmp6.to_csv(path + 'tmp.csv', index = False)
```

```python
tmp = pd.read_csv(path + 'tmp.csv')
print(tmp.shape)
tmp.tail()
```

```python
dataset = tmp.values
dataset = dataset.astype('float32')
dataset
```

```python
scaler = MinMaxScaler(feature_range = (0, 1))
dataset = scaler.fit_transform(dataset)
dataset
```

```python
test_size = 90 * 60
train_size = len(dataset) - test_size
train, test = dataset[:train_size, :], dataset[train_size:len(dataset), :]
print(len(train), len(test))
```

```python
def create_dataset(dataset, look_back=1):
    dataX, dataY = [], []
    for i in range(len(dataset) - look_back - 1):
        a = dataset[i:(i + look_back), 0]
        dataX.append(a)
        dataY.append(dataset[i + look_back, 0])
        
    return np.array(dataX), np.array(dataY)

look_back = 1
trainX, trainY = create_dataset(train, look_back)
testX, testY = create_dataset(test, look_back)
print(trainX.shape, trainY.shape, testX.shape, testY.shape)
```

```python
trainX = np.reshape(trainX, (trainX.shape[0], trainX.shape[1], 1))
testX = np.reshape(testX, (testX.shape[0], testX.shape[1], 1))
print(trainX.shape, testX.shape)
```

```python
model = Sequential()
model.add(LSTM(4, input_shape = (1, look_back)))
model.add(Dense(1))
model.compile(loss = 'mean_squared_error',
              optimizer = 'adam')
model.fit(trainX, trainY, epochs = 100, batch_size = 1,
          verbose = 2)
```

```python
look_back = 3
trainX, trainY = create_dataset(train, look_back)
testX, testY = create_dataset(test, look_back)
print(trainX.shape, trainY.shape, testX.shape, testY.shape)
```

```python
trainX = np.reshape(trainX, (trainX.shape[0], trainX.shape[1], 1))
testX = np.reshape(testX, (testX.shape[0], testX.shape[1], 1))
print(trainX.shape, testX.shape)
```

```python
batch_size = 1
model = Sequential()
model.add(LSTM(4,
               batch_input_shape = (batch_size, look_back, 1),
               stateful = True, return_sequences = True))
model.add(LSTM(4,
               batch_input_shape = (batch_size, look_back, 1),
               stateful = True))
model.add(Dense(1))
model.compile(loss = 'mean_squared_error', optimizer = 'adam')
for i in range(100):
    model.fit(trainX, trainY, epochs = 1,
              batch_size = batch_size, verbose = 2, 
              shuffle = False)
    model.reset_states()
```

```python
batch_size = 1
trainPredict = model.predict(trainX, batch_size = batch_size)
model.reset_states()
testPredict = model.predict(testX, batch_size = batch_size)

trainPredict = scaler.inverse_transform(trainPredict)
trainY = scaler.inverse_transform([trainY])
testPredict = scaler.inverse_transform(testPredict)
testY = scaler.inverse_transform([testY])

trainScore = sqrt(mse(trainY[0], trainPredict[:, 0]))
testScore = sqrt(mse(testY[0], testPredict[:, 0]))
print('Train Score: %.4f RMSE' % (trainScore))
print('Test Score: %.4f RMSE' % (testScore))
```

```python
trainPredictPlot = np.empty_like(dataset)
trainPredictPlot[:, :] = np.nan
trainPredictPlot[look_back:len(trainPredict) + look_back, :] = trainPredict

testPredictPlot = np.empty_like(dataset)
testPredictPlot[:, :] = np.nan
testPredictPlot[(len(trainPredict) + (look_back * 2) + 1):(len(dataset) - 1), :] = testPredict

plt.figure(figsize = (16, 5))
plt.plot(scaler.inverse_transform(dataset), label = 'real')
plt.plot(trainPredictPlot, label = 'train_pred')
plt.plot(testPredictPlot, label = 'test_pred')
plt.legend()
```

# Testing for Stationarity

```python
from statsmodels.tsa.stattools import adfuller

results = adfuller(df_diff)
print(f'p-value: {results[1]}')
```

# Modeling

```python
# Finding the best parameters for ARIMA
from tqdm import tqdm_notebook as tqdm
from itertools import product
from statsmodels.tsa.statespace.sarimax import SARIMAX

def best_param(model, data, pdq, pdqs, exog = None):
    ans = []
    for comb in tqdm(pdq):
        for combs in pdqs:
            try:
                mod = model(data, 
                            order = comb,
                            exog = exog,
                            seasonal_order = combs,
                            enforce_stationarity = False,
                            enforce_invertibility = False,
                            freq = None)
                output = mod.fit()
                ans.append([comb, combs, output.aic])
            except:
                continue
                
    ans_df = pd.DataFrame(ans, 
                          columns = ['pdq', 'pdqs', 'aic'])
    return ans_df.loc[ans_df['aic'] == ans_df['aic'].min()]

p = d = q = range(2)
pdq = list(product(p, d, q))
s = [24]
pdqs = list(product(p, d, q, s))
best_param(SARIMAX, 
           df['ask_price'], 
           pdq,
           pdqs,
           df[df.columns[1]])
```
